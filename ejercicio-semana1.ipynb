```python
def factorial(n):
    if (n == 0 or n == 1):
        return 1
    else:
        return n * factorial(n - 1)


def print_factorial(n):
    resultado = factorial(n)
    print(f"El factorial de {n} es: {resultado}")
```

```python
print_factorial(5)
```

```python
# Hacer una funcion con recursidad para reconocer numeros pares e impares
def esPar(n):
    if n == 0:
        print(f"{n} es un numero par")
    elif n == 1:
        print(f"{n} es un numero impar")
    else:
        return esPar(n - 2)
```

```python
esPar(5)
```

```python
# Funcion para invertir las cadenas
def invertir_cadena(cadena):
    if len(cadena) == 0:
        return cadena
    else:
        return cadena[-1] + invertir_cadena(cadena[:-1])
    #Se esta concatendo el ultimo caracter con la llamada recursiva de la cadena sin el ultimo caracter
    
```

# Rebanadas en Python
Las rebanadas (slices) en Python permiten obtener subcadenas de una cadena. La sintaxis es:

Python

Apply
cadena[inicio:fin:paso]

Inicio es el índice inicial (inclusive).

Fin es el índice final (exclusivo).

Paso es la cantidad en que se avanza (step).

Por ejemplo, para una cadena cadena = "hola", las rebanadas serían:

cadena[:] devuelve toda la cadena.

cadena[::-1] devuelve la cadena invertida.

cadena[0:-1] devuelve "hol", omitiendo el último carácter.

cadena[1:3] devuelve "ol", desde el índice 1 hasta el 2 (exclusivo).

Espero que esta explicación te haya ayudado a entender cómo funciona la función invertir_cadena y los slices en Python.

```python
#Funcion para contar vocales
def contar_vocales(cadena):
    if len(cadena) == 0:
        return 0
    else:
        primera_letra = cadena[0].lower()
        if primera_letra in 'aeiou':
            return 1 + contar_vocales(cadena[1:])
        else:
            return contar_vocales(cadena[1:])
```

```python
vocales = contar_vocales("isla")
print(f"La cantidad de vocales es: {vocales}")
```

```python
set1={1,2,3,4,5}
set2={4,5,6,7,8}
```

```python
#Union
union = set1 | set2
print(f"La union de los dos conjuntos es: {union}")
unionAlter = set1.union(set2)
print(f"La union con metodo alternativode los dos conjuntos es: {unionAlter}")
#Interseccion
interseccion = set1 & set2
print(f"La interseccion de los dos conjuntos es: {interseccion}")
interseccionAlter = set1.intersection(set2)
print(f"La interseccion con metodo alternativo de los dos conjuntos es: {interseccionAlter}")
#Diferencia
diferencia = set1 - set2
print(f"La diferencia de los dos conjuntos es: {diferencia}")
diferenciaAlter = set1.difference(set2)
print(f"La diferencia con metodo alternativo de los dos conjuntos es: {diferenciaAlter}")
```

```python
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Eliminar puntuación
    text = " ".join(text.split())  # Eliminar espacios extra
    return text.lower()


input_text = "     Hello,   World! This is a test.    !!!"
cleaned_text = clean_text(input_text)
print(f"El texto limpio es:'{cleaned_text}'")
```

```python
def is_palindrome(text):
    text = "".join(char.lower() for char in text if char.isalnum())
    return text == text[::-1]


input_text = "oso"#input("Enter a string:")
if is_palindrome(input_text):
    print(f"'{input_text}' is a palíndrome.")
else:
    print(f"'{input_text}' is not a palíndrome.")
```

# Manejo de excepciones

```python
def count_words_and_lines(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            line_count = len(lines)
            word_count = sum(len(line.split())for line in lines)
            print(
                f"El archivo '{filename}' tiene {line_count} líneas y {word_count} palabras.")
    except FileNotFoundError:
        print(f"El archivo '{filename}' no fue encontrado.")


count_words_and_lines("sample.txt")
```

```python
def write_item_to_file(filename, items):
    with open(filename, 'w') as file:
        for item in items:
            file.write(f"{item}\n")


def read_items_from_file(filename):
    try:
        with open(filename, 'r') as file:
            items = file.readlines()
            print(f"Items en el archivo : '{filename}'")
            for item in items:
                print(item.strip())
    except FileNotFoundError:
        print(f"El archivo '{filename}' no fue encontrado.")


read_items_from_file("sample.txt")
```

```python
fruits = ["apple", "banana", "cherry", "date"]
write_item_to_file("fruitsss.txt", fruits)
```

```python
# Funcion map()
numbers = [1, 2, 3, 4] #Declaro el array 
squares_mapped = map(lambda x: x**2,numbers)#numbers pasa como argumento x a la funcion lambda que eleva al cuadrado cada elemento
print(list(squares_mapped))
```

```python
# Funcion filter()
evenList = filter(lambda x: x % 2 == 0, numbers)
print(list(evenList))
```

```python
#  Funcion reduce()
from functools import reduce
product = reduce(lambda x, y: x * y, numbers)
product
```
